# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```

## Activity

Experiment with different parameters on the normal function, and see what effect it has on the shape of the distribution. How does that new shape relate to the standard deviation and variance?

```python
#What happens if we were to change the center point?
incomes = np.random.normal(20, 20, 10000)
plt.hist(incomes, 50)
plt.show()
```

```python
#We can see that the overall shape of the distribution remained the same, but it is shifted to the left on the x-axis
#But what if we changed the standard devation of the sample?
incomes = np.random.normal(100.0, 70.0, 10000)
plt.hist(incomes, 50)
plt.show()
```

```python
#We can see that changing the std deviation does have an effect on the overall shape of the graph
#The peak of the new histogram is a bit different than the original, signifying the mean may have also changed
```
